```python
cat = []
con = []

for k in df.columns:
    if df[k].dtypes == "object":
        cat.append(k)
    else:
        con.append(k)
```

```python
def khu(df):
    cat = list(df.columns[df.dtypes == "object"])

    for khu in df.columns:
        if khu in cat:
            mode = df[khu].mode()[0]
            df[khu] = df[khu].fillna(mode)
        else:
            mean = df[khu].mean()
            df[khu] = df[khu].fillna(mean)
```

```python
khu(df)
```
